# Linear Regression - GD

```python
import numpy as np
import pandas as pd
class linear_regression:

    '''
    >> model = linera_regression()
    >> model.fit(X_train, y_train)
    >> model.preict(X_test)
    '''
    def __init__(self, learning_rate = 0.1, iteration = 100):

        self.learning_rate = learning_rate
        self.iteration = iteration


    def fit(self, X_train, y_train):

        #self.m -> no. of training samples( rows )
        #self.d -> no. of features( cols )
        m, d = X_train.shape

        #initializing Weights(self.W) and bias(self.b) with 0 to start with
        self.W = np.zeros(d)
        self.b = 0

        #list of error to keep track of errors
        self.errors = []

        for i in range(self.iteration):

            #calculating y_hat(prediction) based on initial W, b
            y_hat = self.predict(X_train)

            #calculating partial derivatives/Gradient
            dW = (-1) * np.dot(X_train.T, (y_train - y_hat))/m
            db = (-1) * np.mean(y_train - y_hat)

            #updating weights & bias using the above calculated derivatives/gradients i.e. Gradient Descent
            self.W -= self.learning_rate * dW
            self.b -= self.learning_rate * db

            error = np.mean(np.square( y_train - self.predict(X_train) ))
            self.errors.append(error)

    def predict(self, X):
        return np.dot(X, self.W) + self.b
```

```python
def score(y, y_pred):
    r2 = 1 - (np.sum(np.square(y-y_pred))/np.sum(np.square(y-np.mean(y))))
    return r2
```

# Linear Regression - NE

```python
class LinearRegression_NE:
    def __init__(self):
        self.W = None
        self.b = None

    def fit(self, X_train, y_train):

        self.X_train = X_train

        #self.m -> no. of training samples( rows )
        #self.d -> no. of features( cols )
        m, d = self.X_train.shape

        #concat column of 1 with X_train for bias or w0 term
        self.X_train = pd.concat(( pd.DataFrame(np.ones((m, 1))), self.X_train.reset_index(drop=True)), axis = 1)
        
        # calculate inv(X_t.X).X_t.y
        self.W = pd.DataFrame(np.dot(np.linalg.inv(np.dot(self.X_train.T, self.X_train)),np.dot(self.X_train.T, y_train)))

        self.b = self.W.iloc[0,0]
        self.W = self.W.iloc[1:,:]

    def predict(self, X):
        return np.dot(X, self.W) + self.b
```
